```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
file= 'C:/Users/zoeym/OneDrive/Documents/3rd Sem/ipynb/iris.csv'
cnames = ['sepal_length',
          'sepal_width',
          'petal_length',
          'petal_width',
          'species'] 
iris=pd.read_csv(file,names=cnames)
iris.head(4)
```

```python
iris['species'].value_counts() 
```

# EDA

```python
#Histogram
iris.hist(edgecolor='black',figsize=(10,10), color='orange')   
plt.show()
```

```python
#Bar plot
plt.bar(iris['species'],iris['petal_width'], color='pink')
```

```python
#Scatter plot
iris.plot(kind='scatter', x= 'sepal_length', y='sepal_width', color='red')
plt.show()
```

```python
#2d scatter plot
sns.scatterplot(x="petal_length", y="petal_width",hue='species',data=iris)
```

```python
#Line graph
df = pd.DataFrame(iris,columns=['sepal_length','species'])

plt.plot(df['sepal_length'], df['species'], color='purple') 

plt.title('Sepal Length vs Species',  fontsize=14)
plt.xlabel('sepal_length',  fontsize=14)
plt.ylabel('species',  fontsize=14)
plt.show()
```

```python
#Heat map
params = {'legend.fontsize': 15
          , 'font.size': 8
          , 'font.serif': ['Computer Modern Roman']
          , 'axes.labelsize': 14
          , 'xtick.labelsize': 13
          , 'ytick.labelsize': 13
         }

plt.rcParams.update(params)
corrMatrix = iris.corr(method='pearson')

sns.heatmap(corrMatrix, annot=True, cmap="Oranges")
plt.xticks(rotation=45)
plt.show()
```

```python
#Box plot
iris.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()
```

```python
#Box plot
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.show() 
```

```python
#pairplot
sns.pairplot(iris, hue="species", diag_kind="hist", markers=["o", "s", "D"])
```

```python
#pairplot
sns.pairplot(iris, hue='species',height=3,corner= True)
```

```python
#correlation matrix plot
plt.figure()
sns.heatmap(iris.corr(),annot=True)
plt.show()
```
